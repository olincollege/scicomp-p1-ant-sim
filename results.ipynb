## Replicating Paper Results

I tried to replicate some of the results from the paper, specifically looking at fidelity, deposition rates, and wide/narrow turning kernels.

#### Testing different fidelities
In the paper, they found that decreasing the fidelity lead to the average strength of trails being weaker. While I don't measure a numeric value for strength a possible metric to use is the relative opacity of trails. As the fidelity increases there are more trails which are opaque which would mean those trails are stronger.

```python
%run plot.py --fidelity 0.4 --title "Trail Formation by Ants: Fidelity = 0.4"
%run plot.py --fidelity 0.6 --title "Trail Formation by Ants: Fidelity = 0.6"
%run plot.py --fidelity 0.9 --title "Trail Formation by Ants: Fidelity = 0.9"
```

#### Testing different deposition rates
While I can't measure mine in the same way tht they did since I don't incorporate all of the different variables from the paper, there is a difference between the three plots as the first one has a network which is overall slightly darker and more expansive whereas the second and third progressively have one or two particularly strong trails but aren't as expansive as the first.

```python
%run plot.py --deposition_rate 4 --title "Trail Formation by Ants: Deposition Rate = 4"
%run plot.py --deposition_rate 8 --title "Trail Formation by Ants: Deposition Rate = 8"
%run plot.py --deposition_rate 12 --title "Trail Formation by Ants: Deposition Rate = 12"
```

#### Testing different turning kernels
In the paper, there is a clear difference in the two turning kernels as the narrower kernels resulted in a smaller area being explored. My results do appear to be consistent with that statement as the wide turning kernel has much longer trails than the narrow turning kernel.

```python
%run plot.py --turn_probability 0.56 0.24 0.12 0.06 0.02 --title "Trail Formation by Ants: Wide Turning Kernel"
%run plot.py --turn_probability 0.32 0.3 0.22 0.13 0.03 --title "Trail Formation by Ants: Narrow Turning Kernel"
```

One thing to note is that I'm running most of these with the default value of 1000 iterations whereas the paper runs their simulation with 4000 iterations. It's possible that if I ran more iterations the results would be clearer.
